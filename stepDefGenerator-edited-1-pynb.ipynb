````python
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    tokenizer=tokenizer,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
````

````python
def prompting (code,htmlcode):
    

    prompt = f"""you will be provided with a piece of java code and your task is to remove all the comment lines from it java code:
    {code}"""

    chat = [
       {"role": "system", "content": "you will be provided with a piece of java code and your task is to remove all the comment lines from it."},
       {"role": "user", "content": code}
    ]

    tokenizer.apply_chat_template(chat, tokenize=False)
    
    outputs = pipeline(
        chat,
        max_new_tokens=2000,
    )
    
    output_without_commentLines = outputs[0]["generated_text"][-1]['content']
    print(output_without_commentLines)
    
    """print(output_without_commentLines.split('```')[1])"""
    code = output_without_commentLines.split('```')[1]
    code = f'''{code}'''
    
    chat = [
       {"role": "system", "content": """you will be provided with a piece of java code and/or htmlcode. If in case both java and html codes are provided,
    
            Step 1: Find the relation between java and html code.
            Step 2: For the above code generated the Gherkin feature file, which will be run by the user using cucumber java selenium, 
            Step 3: Add the 'Examples' feature file based on the scenario.
            Step 3: Provide the stepdefinition in java related to the feature already generated in Gherkin with all the necessary packages imported."""}
            ,
       {"role": "user", "content": {code,htmlcode}}
    ]

    tokenizer.apply_chat_template(chat, tokenize=False)
    
    outputs = pipeline(
        chat,
        max_new_tokens=2000,
    )
    stepdefAndFeature = outputs[0]["generated_text"][-1]['content']

    print(stepdefAndFeature)
    stepdef = stepdefAndFeature.split('```')[1]
    stepdef = f'''{stepdef}'''

    feature = stepdefAndFeature.split('```')[3]
    feature = f'''{feature}'''

    print(stepdef)
    print(feature)
````

````python
import gradio as gr

def upload_files(java_file, html_file):
    # Get the contents of the uploaded files
    java_contents = java_file.name
    html_contents = html_file.name

    # Process the file contents
    print("Java file:", java_contents)
    print("HTML file:", html_contents)

    return "Files uploaded successfully!"

demo = gr.Interface(
    fn=upload_files,
    inputs=[gr.File(label="Java file", file_types=["java"]), 
           gr.File(label="HTML file", file_types=["html"])],
    outputs="text",
    title="Upload Files",
    description="Upload a Java file and an HTML file"
)

if __name__ == "__main__":
    demo.launch()
````
